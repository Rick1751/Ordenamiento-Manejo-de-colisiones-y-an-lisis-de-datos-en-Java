```java
import tech.tablesaw.api.*;
import org.knowm.xchart.*;
import java.util.*;

// Leer CSV
Table peliculas = Table.read().csv("movies.csv");

// Top 10 por popularidad
Table top = peliculas.sortDescendingOn("popularity")
                     .first(10);

// Columnas
String[] titulos =
        top.textColumn("title")
           .asObjectArray();

double[] popularidad =
        top.doubleColumn("popularity")
           .asDoubleArray();

// Convertir arrays a listas
List<String> xData = Arrays.asList(titulos);

List<Double> yData = new ArrayList<>();

for (double p : popularidad) {
    yData.add(p);
}

// Crear gráfico
CategoryChart chart = new CategoryChartBuilder()
        .width(1200)
        .height(700)
        .title("Top 10 Películas Más Populares")
        .xAxisTitle("Película")
        .yAxisTitle("Popularidad")
        .build();

// Agregar datos
chart.addSeries("Popularidad", xData, yData);

// Mostrar en Jupyter
BitmapEncoder.getBufferedImage(chart)
```

```java
import tech.tablesaw.api.*;
import java.util.Arrays;

// =======================
// MERGE SORT
// =======================

void mergeSort(Integer[] arr, int left, int right) {

    if (left < right) {

        int mid = (left + right) / 2;

        mergeSort(arr, left, mid);
        mergeSort(arr, mid + 1, right);

        merge(arr, left, mid, right);
    }
}

void merge(Integer[] arr, int left, int mid, int right) {

    int n1 = mid - left + 1;
    int n2 = right - mid;

    Integer[] L = new Integer[n1];
    Integer[] R = new Integer[n2];

    for (int i = 0; i < n1; i++) {
        L[i] = arr[left + i];
    }

    for (int j = 0; j < n2; j++) {
        R[j] = arr[mid + 1 + j];
    }

    int i = 0;
    int j = 0;
    int k = left;

    while (i < n1 && j < n2) {

        if (L[i] <= R[j]) {

            arr[k] = L[i];
            i++;

        } else {

            arr[k] = R[j];
            j++;
        }

        k++;
    }

    while (i < n1) {

        arr[k] = L[i];
        i++;
        k++;
    }

    while (j < n2) {

        arr[k] = R[j];
        j++;
        k++;
    }
}

// =======================
// LEER CSV
// =======================

Table peliculas = Table.read().csv("movies.csv");

Integer[] runtime = peliculas.intColumn("runtime").asObjectArray();

// Ordenar
mergeSort(runtime, 0, runtime.length - 1);

// Mostrar
System.out.println(Arrays.toString(runtime));
```
