```java
import java.nio.file.Files;
import java.nio.file.Paths;
import java.util.ArrayList;
import java.util.List;

// ESTRUCTURA DEL OBJETO PELÍCULA
class Pelicula {
    String titulo;
    int votos;

    public Pelicula(String titulo, int votos) {
        this.titulo = titulo;
        this.votos = votos;
    }

    @Override
    public String toString() {
        return String.format("%s (Votos: %d)", titulo, votos);
    }
}


// MÉTODO QUICKSORT (Mayor a Menor)
void quickSort(Pelicula[] arr, int low, int high) {
    if (low < high) {
        int pi = partition(arr, low, high);
        quickSort(arr, low, pi - 1);
        quickSort(arr, pi + 1, high);
    }
}

void swap(Pelicula[] arr, int i, int j) {
    Pelicula temp = arr[i];
    arr[i] = arr[j];
    arr[j] = temp;
}

int partition(Pelicula[] arr, int low, int high) {
    int mid = low + (high - low) / 2;
    // Orden descendente
    if (arr[low].votos < arr[mid].votos) swap(arr, low, mid);
    if (arr[low].votos < arr[high].votos) swap(arr, low, high);
    if (arr[mid].votos < arr[high].votos) swap(arr, mid, high);
    swap(arr, mid, high);
    
    Pelicula pivot = arr[high]; 
    int i = (low - 1);

    for (int j = low; j < high; j++) {
        if (arr[j].votos >= pivot.votos) {
            i++;
            swap(arr, i, j);
        }
    }
    swap(arr, i + 1, high);
    return i + 1;
}

// EJECUCIÓN

String rutaArchivo = "/workspaces/EDA_Analisis de datos en java/movies.csv"; 

try {
    List<String> lineas = Files.readAllLines(Paths.get(rutaArchivo));
    List<Pelicula> catalogo = new ArrayList<>();

    // k=1 para saltar la cabecera
    for (int k = 1; k < lineas.size(); k++) {
        // Separa por comas respetando los títulos con comillas
        String[] columnas = lineas.get(k).split(",(?=(?:[^\"]*\"[^\"]*\")*[^\"]*$)"); 
        
        try {
            String titulo = columnas[1];
            int votos = Integer.parseInt(columnas[3].trim());
            catalogo.add(new Pelicula(titulo, votos));
        } catch (Exception e) {
            // Ignoramos filas con formato raro
        }
    }

    Pelicula[] peliculasArr = catalogo.toArray(new Pelicula[0]);
    System.out.println(" Procesando " + peliculasArr.length + " películas...\n");

    // Ejecutamos el QuickSort
    quickSort(peliculasArr, 0, peliculasArr.length - 1);

    // Imprimimos el Top 20
    System.out.println(" === LAS 20 PELÍCULAS CON MÁS VOTOS === ");
    for (int i = 0; i < 20; i++) {
        System.out.println((i + 1) + ". " + peliculasArr[i].toString());
    }

    // Exportar TODO al .txt
    String rutaSalida = "/workspaces/Ordenamiento-Manejo-de-colisiones-y-an-lisis-de-datos-en-Java/peliculas_ordenadas.txt";

    StringBuilder sb = new StringBuilder();
    sb.append("===========================================\n");
    sb.append("   PELÍCULAS ORDENADAS DE MAYOR A MENOR VOTO\n");
    sb.append("   Total: ").append(peliculasArr.length).append(" películas\n");
    sb.append("===========================================\n\n");

    for (int i = 0; i < peliculasArr.length; i++) {
        sb.append(String.format("%6d. %s%n", i + 1, peliculasArr[i].toString()));
    }

    Files.write(Paths.get(rutaSalida), sb.toString().getBytes());
    System.out.println(" Archivo generado con " + peliculasArr.length + " películas en:\n   " + rutaSalida);

} catch (Exception e) {
}
```
