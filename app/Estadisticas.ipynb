```java
import tech.tablesaw.api.*;
import org.knowm.xchart.*;
import java.util.*;

// 1. Leer el CSV con Tablesaw
Table peliculas = Table.read().csv("/workspaces/EDA_Analisis de datos en java/movies.csv");

// 2. Forzamos la conversión a DoubleColumn para quitarnos los líos de Int o Long
DoubleColumn colBudget = peliculas.column("budget").asDoubleColumn();
DoubleColumn colRevenue = peliculas.column("revenue").asDoubleColumn();

// Reemplazamos las columnas viejas en la tabla por estas nuevas que sí son Double
peliculas.replaceColumn("budget", colBudget);
peliculas.replaceColumn("revenue", colRevenue);

// 3. Ahora que ambas son fijas DoubleColumn, filtramos sin miedo
Table filtradas = peliculas.where(
    peliculas.doubleColumn("budget").isGreaterThan(0)
    .and(peliculas.doubleColumn("revenue").isGreaterThan(0))
);

// 4. Extraemos los arrays como double[] de una sola vez
double[] budgetArray = filtradas.doubleColumn("budget").asDoubleArray();
double[] revenueArray = filtradas.doubleColumn("revenue").asDoubleArray();

// 5. Convertir a las listas que necesita XChart
List<Double> xData = new ArrayList<>();
List<Double> yData = new ArrayList<>();

for (int i = 0; i < budgetArray.length; i++) {
    xData.add(budgetArray[i]);
    yData.add(revenueArray[i]);
}

// 6. Crear el gráfico tipo Scatter (Dispersión)
XYChart chart = QuickChart.getChart(
    "Presupuesto vs Recaudación", 
    "Presupuesto ($)", 
    "Recaudación ($)", 
    "Películas", 
    xData, 
    yData
);

// 7. Mostrar en Jupyter
BitmapEncoder.getBufferedImage(chart)
```
