# Funciones Hash (Hashing)

Una **función hash** toma una entrada (llave) de tamaño arbitrario y la transforma en un valor de tamaño fijo (hash code) que sirve como índice en una tabla hash para lograr búsquedas y almacenamiento en tiempo eficiente $O(1)$.

---

### 1. Propiedades de una Buena Función Hash
* **Determinista:** Misma entrada siempre genera el mismo hash.
* **Eficiente:** Debe calcularse rápidamente.
* **Uniforme:** Distribuye los valores equitativamente para evitar agrupamientos (clustering).
* **Resistente a Colisiones:** Difícil encontrar dos entradas distintas con el mismo hash.
* **Efecto Avalancha:** Un cambio mínimo en la entrada cambia el hash por completo.

---

### 2. Métodos Comunes de Hashing


| Método | Fórmula / Mecanismo | Pros / Contras |
| :--- | :--- | :--- |
| **División (Módulo)** | $h(k) = k \pmod m$ | **+** Muy simple de implementar.<br>**-** Mala distribución si $m$ no se elige bien. |
| **Multiplicación** | $h(k) = \lfloor m \cdot ((k \cdot A) \pmod 1) \rfloor$ <br> *(Con $0 < A < 1$)* | **+** Menos sensible al tamaño $m$ de la tabla.<br>**-** Computacionalmente más costoso. |
| **Mid-Square** | 1. Elevar la llave al cuadrado ($k^2$).<br>2. Extraer los dígitos centrales. | **+** Buena distribución de los datos.<br>**-** Requiere más procesamiento. |
| **Folding (Plegado)** | 1. Partir la llave en trozos iguales.<br>2. Sumar los trozos (+ módulo si aplica). | **+** Excelente para llaves muy grandes.<br>**-** Depende totalmente de cómo se divida. |
| **Universal** | $h(k) = ((a \cdot k + b) \pmod p) \pmod m$ <br> *(Selecciona función al azar en ejecución)* | **+** Minimiza colisiones ante peores casos.<br>**-** Mayor carga matemática. |
| **Perfecto** | Mapeo único $1:1$ sin colisiones.<br>*(Para un set de datos estático y fijo)* | **+** Cero colisiones.<br>**-** No sirve para datos dinámicos. |
| **Criptográfico** | Algoritmos de seguridad (SHA-256, SHA-3). | **+** Altamente seguro e irreversible.<br>**-** Muy lento para estructuras de datos puras. |

```java
 // Datos aplicar la funcion Hash:
//%: solo devuelve numeros entre o y n-1
        Integer[] arr = vote_count;
        // Tamaño de la tabla hash destino:
        int tamañoHash=100000; // 
        int ans[] = new int[tamañoHash];

        Arrays.fill(ans, -1);
        int hash1 = 0;
        int hashOriginal;

        for (int i = 0; i < ans.length; i++) {
            hash1 = arr[i] % tamañoHash;
            hashOriginal = hash1;
            int j = 1;
            while (ans[hash1] != -1 && j < tamañoHash) {
                hash1 = Math.abs((hashOriginal + (j * j)) % tamañoHash);
                j++;
            }
            ans[hash1] = arr[i];
        }
        System.out.println(Arrays.toString(arr));
        //System.out.println(arr.length);

        System.out.println(Arrays.toString(ans));
        //System.out.println(ans.length);
```
